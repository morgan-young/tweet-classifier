```python
import numpy as np
import pandas as pd
from sklearn import feature_extraction, linear_model, model_selection, preprocessing
```

```python
train_df = pd.read_csv("/kaggle/input/nlp-getting-started/train.csv")
test_df = pd.read_csv("/kaggle/input/nlp-getting-started/test.csv")
```

```python
# Exploring the data
train_df[train_df["target"] == 0]["text"].values[4]
```

```python
train_df[train_df["target"] == 1]["text"].values[20]
```

```python
# Breaking down the tweets into machine readable vectors
count_vectorizer = feature_extraction.text.CountVectorizer()

example_train_vectors = count_vectorizer.fit_transform(train_df["text"][0:5])
```

```python
print(example_train_vectors[0].todense().shape)
print(example_train_vectors[0].todense())
```

```python
# The above lets us know there are 54 unique words in the first 5 tweets.
```

```python
train_vectors = count_vectorizer.fit_transform(train_df["text"])
test_vectors = count_vectorizer.transform(test_df["text"])
```

```python
# Building the classifier
clf = linear_model.SGDClassifier()
```

```python
scores = model_selection.cross_val_score(clf, train_vectors, train_df["target"], cv=3, scoring="f1")
scores
```

```python
# Predicting on the training set
clf.fit(train_vectors, train_df["target"])
```

```python
sample_submission = pd.read_csv("/kaggle/input/nlp-getting-started/sample_submission.csv")
sample_submission["target"] = clf.predict(test_vectors)
sample_submission.head()
```

```python
sample_submission.to_csv("submission.csv", index=False)
```
